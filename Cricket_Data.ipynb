This deliveries can be setup into another new dataframe which will have all the features like the match_id,teams, batting_team, ball_no,batsman, bowler, runs, player_dismissed,city,venue

```python
total_result = pd.DataFrame()


for match_no,match in tqdm(cricket_t20.iterrows()):
    #Initialize a simple and blank dataframe for every match in the innings
    test = pd.DataFrame(columns=[['ball_no','runs','bowler_name','batsman_name','is_wicket']])
    
    current_match = match['innings'][0]['1st innings']['deliveries']
#     print(match)
    
    for enum,delivery in enumerate(current_match):
        ball_no = next(iter(delivery))
        runs = delivery.get(ball_no).get('runs').get('total')
        bowler_name = delivery.get(ball_no).get('bowler')
        batsman_name = delivery.get(ball_no).get('batsman')
        is_wicket = 1 if delivery.get(ball_no).get('wicket') else 0

        test.loc[enum] = [ball_no,runs,bowler_name,batsman_name,is_wicket]

        
    test['match_id'] = match['ID']
    test['teams'] = str(match['info.teams'])
    test['Batting_team'] = match['innings'][0]['1st innings']['team']
    test['city'] = match['info.city']
    test['venue'] = match['info.venue']
    total_result = total_result.append(test)

total_result
```

## Feature Extraction Happens From here

1. Which was the batting team?
2. Which was the bowling team?
3. City where the sport was played
4. What is the current score in the match?
5. How many balls left in the match??
6. What are the number of wickets left?
7. What is the current run rate?
8. What was the status of the last 5 over?

```python
# All the venue names present for the missing city values, have the city name in their stadium

def fetch_city_from_venue(row):
    if pd.isnull(row['city']):
        row['city'] = row['venue'].split()[0]
    return row
```

```python
t20_data = t20_data.apply(fetch_city_from_venue,axis=1)
```

```python
# 4. What is the current score in the match?

t20_data['current_score'] = t20_data.groupby('match_id').cumsum()['runs']
```

```python
t20_data
```

```python
# 4. How many balls left in the match??

t20_data['over'] = t20_data['ball_no'].apply(lambda x: str(x).split('.')[0])
t20_data['ball'] = t20_data['ball_no'].apply(lambda x: str(x).split('.')[1])
```

```python
t20_data.head()
```

```python
t20_data['total_balls_thrown'] = (t20_data['over'].astype('int') * 6) + (t20_data['ball'].astype('int'))
```

```python
t20_data
```

```python
# There are cases where extra balls are thrown (wides,noball etc)  - Eliminating those

t20_data['balls_remaining'] = 120 - t20_data['total_balls_thrown']
```

```python
t20_data.head()
```

```python
t20_data['balls_remaining'] = t20_data['balls_remaining'].apply(lambda x: 0 if x<0 else x)
```

```python
t20_data.head()
```

```python
# 6. What are the number of wickets left?

t20_data['total_player_dismissed'] = t20_data.groupby('match_id').cumsum()['is_wicket']
```

```python
t20_data.head()
```

```python
t20_data['wickets_left'] = 10 - t20_data['total_player_dismissed']
```

```python
t20_data.head()
```

```python
t20_data.drop(columns='total_player_dismissed',inplace=True)
```

```python
# 7. What is the current run rate?

t20_data.head()
```

```python
t20_data['current_rate'] = (t20_data['current_score']*6) / t20_data['total_balls_thrown']
```

```python
t20_data.tail()
```

```python
# 8. What was the status of the last 5 over?

total_matches = t20_data.groupby('match_id')

unique_match_id = t20_data['match_id'].unique()
last_five_overs = []
```

```python
for id in unique_match_id:
    last_five_overs.extend(total_matches.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
t20_data['last_five_over_runs'] = last_five_overs
```

```python
t20_data.head()
```

```python
# Getting the final score of the match

t20_data = t20_data.groupby('match_id').sum()['runs'].reset_index().merge(t20_data, on='match_id')
```

```python
t20_data.head()
```

```python
t20_data_final = t20_data[['Batting_team','Bowling_team','city',\
          'current_score','balls_remaining','wickets_left','current_rate','last_five_over_runs','runs_x']]
```

```python
# To eliminate any bias that might be present

t20_data_final = t20_data_final.sample(frac=1)
```

```python
X_data = t20_data_final.drop(columns='runs_x')
y_data = t20_data_final['runs_x']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_data,y_data,test_size=0.2,random_state=101)
```

```python
X_train
```

```python
# There are three columns named Batting_team, Bowling_team and city which are in string format. 

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trans = ColumnTransformer([
    ('Cat Converter', OneHotEncoder(sparse=False,drop='first'), ['Batting_team', 'Bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor

cricket_pipe = Pipeline(steps=[
    ('Step 1 - Convert cate to num', trans),
    ('Step 2 - Scaling',StandardScaler()),
    ('Step 3 - ', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
cricket_pipe.fit(X_train,y_train)
```

```python
y_pred = cricket_pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(r2_score(y_test, y_pred))
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
pickle.dump(cricket_pipe, open('cricket_pipe_XGB.pkl','wb'))
```
